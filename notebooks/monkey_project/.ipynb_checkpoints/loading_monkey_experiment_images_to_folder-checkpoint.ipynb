```python
import os
import numpy as np
import matplotlib.pyplot as plt

from scipy.misc import face
import torch
import torchvision
import torchvision.transforms as transforms
import gc
```

```python
img_path = "/data/monkey/toliaslab/CSRF19_V1/images/"
images = [np.load(img_path + i) for i in os.listdir(img_path) if ".npy" in i ]
```

```python
# example image
plt.imshow(images[0])
plt.axis("off");
print(images[0])
```

```python
#cropping:
image = images[0]
crop = [(0, 0), (0, 0)]
h, w = image.shape
image = image[crop[0][0]:h - crop[0][1]:1,
                    crop[1][0]:w - crop[1][1]:1]

plt.imshow(image)
```

```python
centerbias = np.zeros(image.shape)
image = np.repeat(image[:, :, np.newaxis], 3, axis=2)

print(image.shape)




```

```python
#cropping
def cropping(image, size):
    crop = size
    h, w = image.shape
    image = image[crop[0][0]:h - crop[0][1]:1,
                    crop[1][0]:w - crop[1][1]:1]
    return image
```

```python
log_dens_array = []
k=0
#for i in img_stack:
for j in range(0,20):
    i = images[j]
    #a = datetime.datetime.now()
    crop = [(20, 20), (20, 20)]
    i = cropping(i, crop)
    image = np.repeat(i[:, :, np.newaxis], 3, axis=2)

    log_dens = deepgaze_2.forward(
        x=torch.tensor([image.transpose(2, 0, 1)], dtype=torch.float32).to('cuda'),
        centerbias=torch.tensor([centerbias]).to('cuda')).clone().detach().cpu()[0].numpy()
    log_dens_array.append(np.exp(log_dens))

    #np.save(str("/data/monkey/toliaslab/CSRF19_V1/images_saliency/individual/" + list_for_saving[k]), log_dens)
    k+= 1
    #b = datetime.datetime.now()
    #print("Process time ", b-a)
print(len(log_dens_array) )   
```
